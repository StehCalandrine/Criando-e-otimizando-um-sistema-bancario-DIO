### Otimizando o sistema bancário com python

```python
usuarios = {}

def criar_conta(usuario, senha):
    if usuario not in usuarios:
        usuarios[usuario] = {"senha": senha, "contas": {}}
        print(f'Conta para o usuário "{usuario}" criada com sucesso!')
    else:
        print('Usuário já existe. Escolha outro nome de usuário.')

def login(usuario, senha):
    if usuario in usuarios and usuarios[usuario]["senha"] == senha:
        print(f'Bem-vindo, {usuario}!')
        return True
    else:
        print('Login falhou. Verifique o nome de usuário e a senha.')
        return False

def criar_nova_conta(usuario, nome_conta):
    if usuario in usuarios:
        nova_conta = {"saldo": 0, "historico": [], "saques_diarios": 3, "limite_saque": 500.0}
        usuarios[usuario]["contas"][nome_conta] = nova_conta
        print(f'Conta "{nome_conta}" criada com sucesso para o usuário "{usuario}"!')
    else:
        print('Usuário não encontrado.')

def realizar_operacao(usuario, nome_conta, operacao, valor):
    if usuario in usuarios and nome_conta in usuarios[usuario]["contas"]:
        conta = usuarios[usuario]["contas"][nome_conta]
        if operacao == "d":
            if valor > 0:
                conta["saldo"] += valor
                conta["historico"].append(f'Depósito: +{valor:.2f}')
                print(f'Depósito de R${valor:.2f} realizado com sucesso na conta "{nome_conta}".')
            else:
                print('O valor do depósito deve ser maior que zero.')
        elif operacao == "s":
            if valor > 0:
                if conta["saldo"] >= valor and conta["saques_diarios"] > 0 and valor <= conta["limite_saque"]:
                    conta["saldo"] -= valor
                    conta["historico"].append(f'Saque: -{valor:.2f}')
                    conta["saques_diarios"] -= 1
                    print(f'Saque de R${valor:.2f} realizado com sucesso na conta "{nome_conta}".')
                elif conta["saques_diarios"] == 0:
                    print('Você atingiu o limite de saques diários.')
                elif valor > conta["limite_saque"]:
                    print(f'O valor do saque excede o limite máximo de R${conta["limite_saque"]:.2f}.')
                else:
                    print('Saldo insuficiente para o saque.')
            else:
                print('O valor do saque deve ser maior que zero.')
        elif operacao == "e":
            print(f'Extrato da conta "{nome_conta}" do usuário "{usuario}":')
            if not conta["historico"]:
                print('Não foram realizadas movimentações.')
            else:
                for item in conta["historico"]:
                    print(item)
            print(f'Saldo atual: R${conta["saldo"]:.2f}')
        else:
            print('Operação inválida. Use "deposito", "saque" ou "extrato".')
    else:
        print('Usuário ou conta não encontrados.')

while True:
    print("\nBem vindo(a)! \nEscolha uma operação:")
    print("1 - Criar Usuário")
    print("2 - Login")
    print("3 - Criar Conta")
    print("4 - Realizar Operação")
    print("5 - Sair")

    escolha = input("Digite o número da operação: ")

    if escolha == "1":
        usuario = input("Digite um nome de usuário: ")
        senha = input("Digite uma senha: ")
        criar_conta(usuario, senha)
    elif escolha == "2":
        usuario = input("Digite o nome de usuário: ")
        senha = input("Digite a senha: ")
        if login(usuario, senha):
            nome_conta = input("Digite o nome da conta: ")
    elif escolha == "3":
        nome_conta = input("Digite o nome da conta: ")
        criar_nova_conta(usuario, nome_conta)
    elif escolha == "4":
        operacao = input("Digite a operação \n [d] Depositar \n [s] Sacar \n [e] Extrato: ")
        valor = float(input("Digite o valor: "))
        realizar_operacao(usuario, nome_conta, operacao, valor)
    elif escolha == "5":
        print("Saindo...")
        break
    else:
        print("Opção inválida. Tente novamente.")
```
